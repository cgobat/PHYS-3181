```python
g_arr = []
x_arr = []
for point in gradient:
    x,y,gx,gy = point
    if x >= -10 and x <= 10 and y==10.15152:
        x_arr.append(x)
        g_arr.append(-gy)
```

```python
steps = np.array([[61, 804], [73, 1134], [85, 1502], [97, 1918], [109, 2402], [121, 2909]])
a1,a2,a3 = np.polyfit(steps[:,0],steps[:,1],2)
stepsx = np.arange(61,242,12)
stepsy = a1*stepsx**2+a2*stepsx+a3
```

```python
#stepsx = np.linspace(61,241,250)
plt.plot(stepsx,stepsy,"--")
plt.plot(stepsx,stepsy,"ok")
plt.xlabel("$N$")
plt.ylabel("Steps to converge ($\delta=10^{-10}$)")
plt.savefig("3.png",dpi=333,bbox_inches="tight")
```
