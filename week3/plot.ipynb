```python
import numpy as np, matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("rk_out.log")
```

```python
X,Y = np.meshgrid(np.linspace(-2.5,2.5,1000),np.linspace(-4,4,1000))
fig,_ = plt.subplots(figsize=(6,8))
plt.contourf(X,Y,X**2*Y**2,levels=300,cmap="plasma")
plt.plot(data[:,1],data[:,2],color="silver")
plt.colorbar(label="Potential")
```
