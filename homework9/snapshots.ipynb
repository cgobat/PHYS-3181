```python
import numpy as np, matplotlib.pyplot as plt
```

```python
X = np.random.normal(10,6,175).tolist()
Y = np.random.normal(10,7,175).tolist()
```

```python
outliersX = np.random.uniform(0,40,25).tolist()
outliersY = np.random.uniform(0,40,25).tolist()
```

```python
plt.figure(figsize=(7,7))
plt.plot(X+outliersX,Y+outliersY,"ko")
plt.xlim(0,40)
plt.ylim(0,40)
plt.savefig("1a.png",dpi=300,bbox_inches="tight")
```

```python
X = np.random.uniform(0,40,200)
Y = np.random.uniform(0,40,200)
plt.figure(figsize=(7,7))
plt.plot(X,Y,"ko")
plt.xlim(0,40)
plt.ylim(0,40)
plt.savefig("1b.png",dpi=300,bbox_inches="tight")
```
