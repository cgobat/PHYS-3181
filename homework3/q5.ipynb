```python
import numpy as np, matplotlib.pyplot as plt
```

```python
orbit = np.genfromtxt("q5orbit.log")
energy = np.genfromtxt("q5energy.log")
```

```python
l1 = plt.plot(energy[:,0],energy[:,1],label=r"$ϵ_k$")
l2 = plt.plot(energy[:,0],energy[:,2],label=r"$ϵ_p$")
l3 = plt.plot(energy[:,0],energy[:,1]+energy[:,2],label=r"$ϵ_k+ϵ_p$")
plt.ylabel("Specific orbital energy ($ϵ$)")
plt.ylim(-0.2,0.2)
plt.xlim(0,30000)
plt.xlabel("Time [yr]")
ax2 = plt.twinx()
l4 = ax2.plot(orbit[:,0],np.sqrt(orbit[:,1]**2+orbit[:,2]**2),color="purple",label="$R$")
ax2.set_ylabel("Distance to Sun ($R$) [AU]")
lines = l1+l2+l3+l4
labs = [l.get_label() for l in lines]
plt.legend(lines,labs, loc="upper left",bbox_to_anchor=(1.2,1))
plt.savefig("q5energy.pdf",bbox_inches="tight",dpi=300)
```
